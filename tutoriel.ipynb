# Tutoriel filtrage de Kalman

```python
import numpy as np
from filtrages_kalman import script_classic_trajectory, get_err
import  matplotlib.pylab as plt
%matplotlib inline
```

```python
real_path_point, l_pos_filtre, l_pos_measured = script_classic_trajectory(10)
```

```python
def get_x_array(l):
    return np.array([point.x for point in l])

def get_y_array(l):
    return np.array([point.y for point in l])

def plot_real_measured_filtered(val=10):
    real_path_point, l_pos_filtre, l_pos_measured = script_classic_trajectory(val)
    f, (ax1, ax2, ax3) = plt.subplots(3, 1, sharey=True, figsize=(8, 14))
    ax1.plot(get_x_array(real_path_point), get_y_array(real_path_point))
    ax1.set_title("Comparaison de l'observation brute et de l'observation bruite")
    ax2.plot(get_x_array(l_pos_measured), get_y_array(l_pos_measured))
    ax3.plot(get_x_array(l_pos_filtre), get_y_array(l_pos_filtre))
plot_real_measured_filtered(10)
plot_real_measured_filtered(20)
plot_real_measured_filtered(50)
plot_real_measured_filtered(100)
```

```python
x = np.array([point.x for point in l_pos_measured])
y = np.array([point.y for point in l_pos_measured])
plt.plot(x, y)
```

```python
x = np.array([point.x for point in l_pos_filtre])
y = np.array([point.y for point in l_pos_filtre])
plt.plot(x, y)
```

```python
print("Ecarts entre le chemin rel et le chemin mesur")
y_observed = get_err(real_path_point, l_pos_measured)
plt.plot(x, y_observed)
```

```python
print("Ecarts entre le chemin rel et le chemin filtr")
y_filtered = get_err(real_path_point, l_pos_filtre)
plt.plot(x, y_filtered)
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y_observed)
ax1.set_title("Comparaison de l'observation brute et de l'observation bruite")
ax2.plot(x, y_filtered)
```
